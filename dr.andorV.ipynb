```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\anton\Downloads\crap_dupliacates.csv")
df
```

```python
df['visit_duration'] = pd.to_timedelta(df['visit_duration'])

df['visit_duration'] = df['visit_duration'].dt.total_seconds()

df = df[df["visit_duration"] != 0]
```

```python
condition_1 = df["Variation"] == "Test"
condition_2 = df["Variation"] == "Control"
test_= df[condition_1]["visit_duration"]
control_ = df[condition_2]["visit_duration"]

#h0: avg test = avg control
#h1: avg test != avg control

stats.ttest_ind(test_, control_)

#p>alpha, so we reject null hypothesis
```
